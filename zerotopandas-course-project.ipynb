````python
Generation_count.plot.pie(y='Name',figsize=(5,5))
labels=['Gen1','Gen2','Gen3','Gen4','Gen5','Gen6']
plt.legend(labels)
jovian.commit() # plotting it on pie chart
````

#### Q3: Find top 10 attackers of all types

#### Q4:  Find the top 10 defenders of all types

#### Q5: Analyse the population of pokemon of all type

````python
df_3=df_1['type1'].value_counts()
labels=list(df_3.index)
colors =['red','yellow','grey','orange','lightskyblue','green','gold','hotpink', 'b', '#4CAF50',
        '#00ffff','#00bfff','#bf4040','#ffcccc','#ffd966','#6666ff','#bfd98c','#ccaf66']
plt.pie(df_3,labels= labels,colors=colors,autopct= "%1.1f%%")
plt.legend(labels)
plt.title("Type 1 of Pokemons", fontsize=12)

plt.show()

df_3=df_1['type2'].value_counts()
labels=list(df_3.index)
colors =['red','yellow','grey','black','lightskyblue','green','gold','hotpink', 'b', '#4CAF50',
        '#00ffff','#00bfff','#bf4040','#ffcccc','#ffd966','#6666ff','#bfd98c','#ccaf66']
plt.pie(df_3,labels= labels,colors=colors,autopct= "%1.1f%%")
plt.legend(labels)
plt.title("Type 2 of Pokemons", fontsize=12)

plt.show()
````

````python
plt.rcParams["figure.figsize"] = (20,20)
plt.rcParams['font.size'] = 20   
sns.countplot(x ='type1',data=pokemon_data)  
plt.title('Type 1 population of pokemon')
plt.rcParams['font.size'] = 20 

````

````python
plt.rcParams["figure.figsize"] = (20,20)
plt.rcParams['font.size'] = 20   
sns.countplot(x ='type2',data=pokemon_data)  
plt.title('Type 2  population of pokemon')
plt.rcParams['font.size'] = 20
````

#### Water has the highest population in type1 with (14%),flying is the majority in type 2 with (23.4%) population,
ground, ghost and dragon have same type of population in type1(4%), normal/flying is the most common population for dual types
ground,poison,psychic has 2nd most populaiton in type2
